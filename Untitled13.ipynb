```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from torchvision.datasets import VOCDetection
from torch.utils.data import DataLoader, Dataset
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from torchvision.ops import box_iou

# --- Detection Head ---
class YOLOHead(nn.Module):
    def __init__(self, in_channels, num_classes=20):
        super(YOLOHead, self).__init__()
        self.pred = nn.Conv2d(in_channels, 5 + num_classes, kernel_size=1)

    def forward(self, x):
        return self.pred(x)

# --- Dataset Wrapper ---
class VOCDataset(Dataset):
    def __init__(self, root, year='2007', image_set='train', transform=None):
        self.dataset = VOCDetection(root=root, year=year, image_set=image_set, download=True)
        self.transform = transform

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, target = self.dataset[idx]
        w, h = img.size
        boxes, labels = [], []

        for obj in target['annotation']['object']:
            bbox = obj['bndbox']
            xmin = int(bbox['xmin']) / w
            xmax = int(bbox['xmax']) / w
            ymin = int(bbox['ymin']) / h
            ymax = int(bbox['ymax']) / h
            xc = (xmin + xmax) / 2
            yc = (ymin + ymax) / 2
            bw = xmax - xmin
            bh = ymax - ymin
            boxes.append([xc, yc, bw, bh])
            labels.append(1)

        boxes = torch.tensor(boxes, dtype=torch.float32)
        labels = torch.tensor(labels, dtype=torch.float32)

        if self.transform:
            img = self.transform(img)

        targets = torch.cat([boxes, labels.unsqueeze(1)], dim=1)
        return img, targets

def collate_fn(batch):
    imgs, targets = zip(*batch)
    return list(imgs), list(targets)

# --- Model with MobileNet Backbone ---
class YOLOMobileNet(nn.Module):
    def __init__(self, num_classes=20):
        super(YOLOMobileNet, self).__init__()
        mobilenet = models.mobilenet_v2(weights='IMAGENET1K_V1')
        self.backbone = mobilenet.features
        self.head = YOLOHead(1280, num_classes)

    def forward(self, x):
        x = self.backbone(x)
        x = self.head(x)
        return x

# --- YOLO to XYXY Converter ---
def yolo_to_xyxy(box, img_width=224, img_height=224):
    x_c, y_c, w, h = box
    x1 = (x_c - w / 2.0) * img_width
    y1 = (y_c - h / 2.0) * img_height
    x2 = (x_c + w / 2.0) * img_width
    y2 = (y_c + h / 2.0) * img_height
    return [x1, y1, x2, y2]

# --- Loss Function (Simple Placeholder) ---
def detection_loss(preds, targets):
    return ((preds - 0.5) ** 2).mean()

# --- Train Model ---
def train_model():
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])
    dataset = VOCDataset(root='data/', year='2007', image_set='train', transform=transform)
    dataloader = DataLoader(dataset, batch_size=4, shuffle=True, collate_fn=collate_fn)

    model = YOLOMobileNet(num_classes=20)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    criterion = detection_loss

    for epoch in range(5):
        model.train()
        total_loss = 0
        for imgs, _ in dataloader:
            imgs = torch.stack(imgs).to(device)
            preds = model(imgs)
            loss = criterion(preds, torch.zeros_like(preds))
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1}, Loss: {total_loss:.4f}")

    torch.save(model.state_dict(), "yolo_mobilenet_voc.pth")
    return model

```

```python
# --- Inference ---
def run_inference(model_path='yolo_mobilenet_voc.pth'):
    model = YOLOMobileNet()
    model.load_state_dict(torch.load(model_path, map_location=torch.device('cpu')))
    model.eval()

    image = Image.open("demo.jpg").convert("RGB")
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])
    img_tensor = transform(image).unsqueeze(0)
    output = model(img_tensor).squeeze(0).permute(1, 2, 0).detach().numpy()
    boxes = output.reshape(-1, 25)
    boxes = boxes[np.max(boxes[:, 5:], axis=1) > 0.3]

    fig, ax = plt.subplots(1)
    ax.imshow(image)
    for b in boxes:
        x1, y1, x2, y2 = yolo_to_xyxy(b[:4], image.width, image.height)
        rect = plt.Rectangle((x1, y1), x2 - x1, y2 - y1, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)
    plt.show()

# --- Evaluate Model ---
def evaluate_model(model, iou_threshold=0.5, conf_threshold=0.3):
    model.eval()
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
    ])
    val_dataset = VOCDataset(root='data/', year='2007', image_set='val', transform=transform)
    val_loader = DataLoader(val_dataset, batch_size=1, shuffle=False, collate_fn=collate_fn)

    total_tp, total_fp, total_fn = 0, 0, 0
    iou_scores = []

    with torch.no_grad():
        for images, targets in val_loader:
            images = torch.stack(images)
            outputs = model(images).squeeze(0).permute(1, 2, 0).reshape(-1, 25)
            probs = torch.sigmoid(outputs[:, 4])
            preds = outputs[probs > conf_threshold]

            pred_boxes = [yolo_to_xyxy(p[:4].cpu().numpy()) for p in preds]
            gt_boxes = [yolo_to_xyxy(t[:4].numpy()) for t in targets[0]]

            if not pred_boxes:
                total_fn += len(gt_boxes)
                continue

            iou_matrix = box_iou(torch.tensor(pred_boxes), torch.tensor(gt_boxes))
            max_iou, _ = iou_matrix.max(dim=1)
            tp = (max_iou > iou_threshold).sum().item()
            fp = len(pred_boxes) - tp
            fn = len(gt_boxes) - tp

            total_tp += tp
            total_fp += fp
            total_fn += fn
            iou_scores.extend(max_iou[max_iou > iou_threshold].cpu().numpy().tolist())

    precision = total_tp / (total_tp + total_fp + 1e-6)
    recall = total_tp / (total_tp + total_fn + 1e-6)
    mean_iou = np.mean(iou_scores) if iou_scores else 0.0

    print("\n--- Evaluation Metrics ---")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"mIoU: {mean_iou:.4f}")
    return precision, recall, mean_iou
```

```python

# --- Main Entry ---
if __name__ == "__main__":
    model = train_model()
    run_inference()
    evaluate_model(model)
```
